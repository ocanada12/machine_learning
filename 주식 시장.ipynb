```python
import pandas as pd
from urllib.request import urlopen
from bs4 import BeautifulSoup
import numpy as np
```

대만지수

```python
path = "https://kr.tradingview.com/markets/stocks-taiwan/market-movers-large-cap/"
html = urlopen(path)  
bsObject = BeautifulSoup(html, "html.parser") 

address = "https://finance.naver.com/marketindex/exchangeDetail.nhn?marketindexCd=FX_TWDKRW"
add = urlopen(address)  
ad = BeautifulSoup(add, "html.parser") 

currency = float(ad("tr")[1]("td")[0].text)

taiwan_market = pd.DataFrame()
taiwan_market_temp = pd.DataFrame()
taiwan_market_temp = taiwan_market_temp.append({"종목":"","시가총액":""}, ignore_index=True)

for i in range(0,100):
    value = round(float(bsObject("tbody",{"class":"tv-data-table__tbody"}
            )[0]("tr",{"class":"tv-data-table__row tv-data-table__stroke tv-screener-table__result-row"}
                )[i]("td", {"class":"tv-data-table__cell tv-screener-table__cell tv-screener-table__cell--big"}
                    )[-4].text[:-1])*currency/1000,2)
    name = bsObject("tbody", {"class":"tv-data-table__tbody"}
                                           )[0]("span", {"class":"tv-screener__description"})[i].text[10:]
    if value != 0 :
        taiwan_market_temp.iat[0,1] = name
        taiwan_market_temp.iat[0,0] = value
        taiwan_market= taiwan_market.append(taiwan_market_temp,ignore_index=True)
    else :
        pass
taiwan_market = taiwan_market[["종목","시가총액"]]
```

```python
import pandas_datareader as pdr
from datetime import datetime
start = "2020-01-20"
end = "2020-01-21"
Hon = pdr.DataReader('2317.TW', 'yahoo', start, end)
Tsm = pdr.DataReader('2330.TW', 'yahoo', start, end)
For = pdr.DataReader('6505.TW', 'yahoo', start, end)
Ap = pdr.DataReader('AAPL', 'yahoo', start, end)
Sam = pdr.DataReader('005930.KS', 'yahoo', start, end)

tot = pdr.DataReader('7203.T', 'yahoo', start, end)
lg = pdr.DataReader('066570.KS', 'yahoo', start, end)



```

```python
Hon
```

```python
Tsm["market cap"] = Tsm["Close"]*(taiwan_market.iat[0,1]/Tsm["Close"].iat[-1])
Hon["market cap"] = Hon["Close"]*(taiwan_market.iat[1,1]/Hon["Close"].iat[-1])
For["market cap"] = For["Close"]*(taiwan_market.iat[2,1]/For["Close"].iat[-1])
Sam["market cap"] = Sam["Close"]*(korea_market.iat[0,1]/Sam["Close"].iat[-1])
Ap["market cap"] = Ap["Close"]*(usa_market.iat[0,1]/Ap["Close"].iat[-1])
tot["market cap"] = tot["Close"]*(japan_market.iat[0,1]/tot["Close"].iat[-1])
```

```python
Tsm["market cap"] = Tsm["Close"]/Tsm["Close"].iat[0]
Hon["market cap"] = Hon["Close"]/Hon["Close"].iat[0]
For["market cap"] = For["Close"]/For["Close"].iat[0]
Sam["market cap"] = Sam["Close"]/Sam["Close"].iat[0]
Ap["market cap"] = Ap["Close"]/Ap["Close"].iat[0]
tot["market cap"] = tot["Close"]/tot["Close"].iat[0]
lg["market cap"] = lg["Close"]/lg["Close"].iat[0]
```

```python
from matplotlib import pyplot as plt
plt.plot(Hon.index,Hon["market cap"], color = "blue")
plt.plot(Sam.index,Sam["market cap"], color = "green")
plt.plot(Tsm.index,Tsm["market cap"], color = "purple")
plt.plot(tot.index,tot["market cap"], color = "orange")

```

```python
import pandas_datareader as pdr
from datetime import datetime
start = "1980-01-20"
end = "2020-01-20"
korea = pdr.DataReader('^KS11', 'yahoo', start, end)
taiwan = pdr.DataReader('^TWII', 'yahoo', start, end)
hongkong = pdr.DataReader('^HSI', 'yahoo', start, end)
china = pdr.DataReader('000001.SS', 'yahoo', start, end)
euro = pdr.DataReader('^N100', 'yahoo', start, end)
japan = pdr.DataReader('^N225', 'yahoo', start, end)
indonesia = pdr.DataReader('^JKSE', 'yahoo', start, end)
russia = pdr.DataReader('IMOEX.ME', 'yahoo', start, end)

```

```python
korea["market cap"] = korea["Close"]/korea["Close"].iat[0]
taiwan["market cap"] = taiwan["Close"]/taiwan["Close"].iat[0]
hongkong["market cap"] = hongkong["Close"]/hongkong["Close"].iat[0]
china["market cap"] = china["Close"]/china["Close"].iat[0]
euro["market cap"] = euro["Close"]/euro["Close"].iat[0]
japan["market cap"] = japan["Close"]/japan["Close"].iat[0]
indonesia["market cap"] = indonesia["Close"]/indonesia["Close"].iat[0]
russia["market cap"] = russia["Close"]/russia["Close"].iat[0]
```

```python
from matplotlib import pyplot as plt
plt.plot(korea.index,korea["market cap"], color = "blue")
plt.plot(taiwan.index,taiwan["market cap"], color = "green")
plt.plot(euro.index,euro["market cap"], color = "pink")
plt.plot(russia.index,russia["market cap"], color = "aqua")
```
