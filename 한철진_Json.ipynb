```python
movieIDDF = pd.DataFrame() #아무것도 없는 데이터프레임 만들기
movieIDDF_temp = pd.DataFrame()
movieIDDF_temp = movieIDDF_temp.append({"directors":"","openDt":"","genreAlt":"","movieCd":"","movieNm":"","movieNmEn":""}, ignore_index=True)
```

```python
for i in range(0,length):
    
        if len(movieIdData['movieListResult']['movieList'][i]["directors"]) != 0 :
            movieIDDF_temp.iat[0,0] = movieIdData['movieListResult']['movieList'][i]["directors"][0]["peopleNm"]
            movieIDDF_temp.iat[0,1] = movieIdData['movieListResult']['movieList'][i]["genreAlt"]
            movieIDDF_temp.iat[0,2] = movieIdData['movieListResult']['movieList'][i]["movieCd"]
            movieIDDF_temp.iat[0,3] = movieIdData['movieListResult']['movieList'][i]["movieNm"]
            movieIDDF_temp.iat[0,4] = movieIdData['movieListResult']['movieList'][i]["movieNmEn"]
            movieIDDF_temp.iat[0,5] = movieIdData['movieListResult']['movieList'][i]["openDt"]
        else:
            movieIDDF_temp.iat[0,0] = ""
            movieIDDF_temp.iat[0,1] = movieIdData['movieListResult']['movieList'][i]["genreAlt"]
            movieIDDF_temp.iat[0,2] = movieIdData['movieListResult']['movieList'][i]["movieCd"]
            movieIDDF_temp.iat[0,3] = movieIdData['movieListResult']['movieList'][i]["movieNm"]
            movieIDDF_temp.iat[0,4] = movieIdData['movieListResult']['movieList'][i]["movieNmEn"]
            movieIDDF_temp.iat[0,5] = movieIdData['movieListResult']['movieList'][i]["openDt"]
    
        movieIDDF= movieIDDF.append(movieIDDF_temp,ignore_index=True)
```

```python
movieIDDF["year"] = movieIDDF["openDt"].apply(lambda x: x[0:4])
movieIDDF["month"] = movieIDDF["openDt"].apply(lambda x: x[4:6])
movieIDDF["day"] = movieIDDF["openDt"].apply(lambda x: x[6:])
```

```python
movieIDDF = movieIDDF.sort_values(by="openDt")
movieIDDF = movieIDDF.set_index("movieCd")
```
